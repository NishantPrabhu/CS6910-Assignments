```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

from models import Network
from layers import Dense, Input 
from losses import RMSE 
from activations import Sigmoid, ReLU, Tanh, Softmax, Linear
from optimizers import SGD, AdaGrad, AdaDelta, Adam
from utils import to_categorical
from plotting import SpatialPlot, ParityPlot
```

```python
train_clf = pd.read_csv(".././data/2d_nonlinear/2d_nonlinear_data.csv")
train_clf = train_clf[train_clf['label']==0]

X_train = train_clf[['x1', 'x2']].values.reshape((-1, 2))
y_true = train_clf['label'].values.reshape((-1, 1))
y_train = to_categorical(y_true)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
```

```python
layers = [
    Input(2, label='Input'),
    Dense(10, activation=Sigmoid(), label='Hidden_1'),
    Dense(10, activation=Sigmoid(), label='Hidden_2'),
    Dense(1, activation=Softmax(), label='Output')
]

model = Network()
model.set_layers(layers)
model.compile(loss='CrossEntropy', optimizer=SGD(lr=1.0, momentum=0.8), metric='Accuracy')
```

```python
model.train(X_train, y_train, epochs=500, log_frequency=100)
```

```python
y_probs = model.predict(X_train)
y_pred = np.array([np.argmax(i) for i in y_probs])
```

```python
SpatialPlot(X_train, y_true, y_pred)
```
