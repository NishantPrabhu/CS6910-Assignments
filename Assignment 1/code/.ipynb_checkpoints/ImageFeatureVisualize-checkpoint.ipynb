```python
import numpy as np
from utils import ImageFeatureExtractor, OneHotEncoder
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Load features
save_path = '.././data/image_data/FeatureExtraction_2D'
X_train, y_true, label_map = ImageFeatureExtractor().load_features(save_path)
y_train = OneHotEncoder().transform(y_true)
```

```python
# PCA of features
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)

plt.figure(figsize=(8, 8))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_true, s=20)
plt.title("PCA transformed : Scatter")
plt.show()
```

```python
# PCA of features
tsne = TSNE(n_components=2)
X_train_tsne = tsne.fit_transform(X_train)

plt.figure(figsize=(8, 8))
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1], c=y_true, s=20)
plt.title("TSNE transformed : Scatter")
plt.show()
```
